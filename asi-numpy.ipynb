5.write a NumPy program to compute the sum of the diagonal element of a given array(matrix).

```python
from numpy import *
arr1=[[1,2,3],[4,5,6],[9,9,1]]
arr2=[[3,2,1],[6,3,4],[3,1,1]]
a=diagonal(arr1)
b=diagonal(arr2)
print("First array")
for i in range(0,len(arr1)):
    print(arr1[i])
print("second array")
for j in range(0,len(arr2)):
    print(arr2[j])
    
print("The diagonal of first array:",a)
print("The diagonal of second array:",b)
c=a+b
print("The sum of diagonal elements",c)
```

4.Write a NumPy program to compute the condition number of a given matrix.

[3 4]]
8.Condition number of the said matrix:
9.14.9330343737

```python
import numpy as np
import numpy.linalg as a1
arr1=np.array([[1,2],[3,4]])
print("Original matrix")
print(arr1)
print("Condition number of the said matrix")
print(a1.cond(arr1)) 
```
